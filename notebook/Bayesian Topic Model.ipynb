```python
# Load the corpus and dictionary
from gensim import corpora, models
import pyLDAvis.gensim

corpus = corpora.MmCorpus('../raw_corpus2.mm')
dictionary = corpora.Dictionary.load('../obama.dic')
```

```python
lda = models.LdaModel.load('../obama25.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(followers_data)
```
